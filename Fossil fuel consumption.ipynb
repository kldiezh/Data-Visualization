#### Importing libraries for data processing and visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Loading data

```python
df = pd.read_csv('global_fossil_fuel_consumption.csv')
df.head()
```

```python
# Rename columns
df.rename(columns={'Gas (TWh, direct energy)':'Gas', 'Oil (TWh, direct energy)':'Oil', 'Coal (TWh, direct energy)':'Coal'}, inplace=True)
df.head()
```

#### Bar Plot

```python
df_filtered = df[df['Year'] >= 2010] # Filter data for years from 2000 onwards

df_filtered.set_index('Year', inplace=True) # Set the 'Year' column as the index

ax = df_filtered.plot(kind = 'bar', title='Global fossil fuel consumption', xlabel='Years', ylabel='TWh')

ax.legend(loc='upper left', bbox_to_anchor=(1, 1)) # Place the legend outside the plot

plt.show()
```

#### Area Plot

```python
df.set_index('Year', inplace=True)

bx = df.plot.area(title='Global fossil fuel consumption', xlabel='Years', ylabel='Twh')

bx.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.show()
```

#### Line Plot

```python
cx = df.plot.line(title='Global fossil fuel consumption', xlabel='Years', ylabel='Twh')

cx.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.show()
```
