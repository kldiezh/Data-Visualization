## **Import libraries**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme(style="darkgrid")
```

## **Dataset Exploration**

```python
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
```

```python
df.info()
```

```python
columnas_categoricas = df.select_dtypes(include=['object', 'category'])
```

## **Dataset Visualization**

**Histograms for Numerical Columns**

```python
numerical_cols = ['Age', 'Sleep Duration', 'Quality of Sleep', 
                  'Physical Activity Level', 'Stress Level', 
                  'Heart Rate', 'Daily Steps']

df[numerical_cols].hist(bins=12, figsize=(12, 10))
plt.suptitle('Distributions of Numerical Variables', fontsize=14)
plt.tight_layout()
plt.show()
```

**Boxplot**

```python
sns.boxplot(x='BMI Category', y='Stress Level', data=df)
```

**Categorical vs Numerical: Barplots**

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Occupation', y='Sleep Duration', data=df, estimator='mean')
plt.title('Average Sleep Duration by Occupation')
plt.xticks(rotation=90)
plt.show()
```

**Grouped Barplot**

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Sleep Disorder', y='Stress Level', hue='Gender', data=df)
plt.title('Stress Level by Sleep Disorder and Gender')
plt.show()
```

**Displot**

```python
sns.displot(df, x='Physical Activity Level', hue ='Sleep Disorder', multiple='stack')
```
